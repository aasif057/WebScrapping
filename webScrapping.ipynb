```python
import requests
from bs4 import BeautifulSoup as Bs
import pandas as pd
```

```python
r = requests.get("https://www.nulearn.in/")
c = r.content
soup = Bs(c,"html.parser")
links = []

for link in soup.find_all('a', {'class':'mega-nav-link'}):
    a = link.get('href')
    links.append(a)
links = links[: len(links)-3]
status = []
for link in soup.find_all('a', {'class':'mega-nav-link'}):
    b = link.find_all('span')
    status.append(b[0].select('i')[0].text.strip())
    if len(status)==len(links):
        break
```

```python
course_name = []
duration = []

institution = []
description=[]
about=[]
fee_ind = []
fee_foreign = []
for i in range(len(links)):
    r = requests.get(links[i])
    c = r.content
    soup = Bs(c,'html.parser')
    course_name.append(soup.select('h1.course-page-name')[0].text.strip())
    institution.append(soup.select('h4.course-banner-sub-lael')[0].text.strip())
    duration.append(soup.select('h4.course-banner-sub-lael')[1].text.strip())
    description.append(soup.select('p.course-banner-para')[0].text.strip())
    about.append(soup.select('p')[1].text.strip())
    fee_ind.append(soup.select('h2.program-fee-amount')[0].text.strip())
    fee_foreign.append(soup.select('h2.program-fee-amount')[1].text.strip())
    
```

```python
fee_foreign[26] = soup.select('h2.program-fee-amount')[2].text.strip()
fee_foreign[23] = soup.select('h2.program-fee-amount')[2].text.strip()
fee_foreign
```

```python
features = []
for i in range(len(links)):
    r = requests.get(links[i])
    c = r.content
    soup = Bs(c,'html.parser')
    
    for link in soup.find_all('div', {'class':'highlight-block-wrapper'}):
        f = []
        for i in range(5):
            f.append(link.select('div.highlightlabel')[i].text.strip())
    features.append(f)
```

```python
dict = {'Course Name':course_name,'Institution Name':institution,'Status':status,'Course Duration':duration,'Fee for Indian Students':fee_ind,'Fee for International Students':fee_foreign,'Description of the Course':description,'About the Course':about,'Key Features':features}
df = pd.DataFrame(dict)
```
